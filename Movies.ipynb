```python
# 6 Filter data(Rating<9)
import pandas as pd
df=pd.read_csv("Movies.csv")
filtered=df[df["IMDB_Rating"]<9]
print(filtered)
```

```python
# 7 Find missing values
import pandas as pd
df = pd.read_csv("Movies.csv")
print(df.isnull())
print(df.isnull().sum())
```

```python
# 8 Fill missing values
import pandas as pd
df = pd.read_csv("Movies.csv")
df["IMDB_Rating"] = df["IMDB_Rating"].fillna(df["IMDB_Rating"].mean())
print(df)
```

```python
# 9 Remove duplicates rows
import pandas as pd
df = pd.read_csv("Movies.csv")
df=df.drop_duplicates()
print(df)
```

```python
# 10 Simple Movie rating Analysis()
import pandas as pd
df = pd.read_csv("Movies.csv")
print("Highest Rating:",df["IMDB_Rating"].max())
print("Average Rating:",df["IMDB_Rating"].mean())
print("Lower Rating:",df["IMDB_Rating"].min())
```
